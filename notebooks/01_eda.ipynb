# DermAssist-AI — Exploratory Data Analysis
Visualise class distribution, sample images, and pixel statistics.

```python
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import random

DATA = Path('../data/processed/train')
classes = sorted([d.name for d in DATA.iterdir() if d.is_dir()])
counts  = {c: len(list((DATA/c).glob('*'))) for c in classes}

plt.bar(counts.keys(), counts.values())
plt.title('Class distribution (train)')
plt.ylabel('# images')
plt.show()
print(counts)
```

```python
fig, axes = plt.subplots(2, 4, figsize=(14, 6))
for ax, cls in zip(axes.flatten(), classes * 2):
    imgs = list((DATA/cls).glob('*'))
    if imgs:
        ax.imshow(Image.open(random.choice(imgs)))
        ax.set_title(cls)
        ax.axis('off')
plt.tight_layout()
plt.show()
```
